```python
def standardize_gender(g):
    g = str(g).strip().upper()
    if g in ['M', 'MALE']:
        return 'Male'
    elif g in ['F', 'FEMALE']:
        return 'Female'
    elif g == 'OTHER':
        return 'Other'
    else:
        return g

df['gender'] = df['gender'].apply(standardize_gender)
```

```python
trustworthy = df[~df['order_id'].isin(['ORD-00539','ORD-01025','ORD-00809','ORD-02560'])]
median_margin = (trustworthy['profit'] / trustworthy['sales_amount']).median()
print(median_margin)
```

```python
corrupted_ids = ['ORD-00539', 'ORD-01025', 'ORD-00809', 'ORD-02560']
mask = df['order_id'].isin(corrupted_ids)

df['sales_amount_recalculated'] = mask

df.loc[mask, 'sales_amount'] = df.loc[mask, 'quantity'] * df.loc[mask, 'unit_price'] * (1 - df.loc[mask, 'discount_pct'])
df.loc[mask, 'profit'] = df.loc[mask, 'sales_amount'] * median_margin
```
